```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.naive_bayes import GaussianNB

SVM = svm.SVC(gamma='scale',class_weight='balanced',random_state=42)
lsvm = svm.SVC(kernel='linear',class_weight='balanced',random_state=42)
knn = KNeighborsClassifier(n_neighbors=5, weights='distance')
lr = LogisticRegression(class_weight='balanced',random_state=42)
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(random_state=42)
rfb = RandomForestClassifier(random_state=42,class_weight='balanced_subsample')
ab = AdaBoostClassifier(random_state=42)
nb = GaussianNB()
```

```python
result = pd.DataFrame(columns=['accuracy', 'precision', 'recall','specifity','f1'])
nb = GaussianNB()
#nb.fit(SMOTE_train_x, SMOTE_train_y) 
nb.fit(X_ss_train, y_train)
y_pred_nb = nb.predict(X_ss_test)
accuracy=accuracy_score(y_test, y_pred_nb)
print(accuracy)
print(confusion_matrix(y_test, y_pred_nb))
print('-----------------------------------------------------------')
print(classification_report(y_test, y_pred_nb))
scores = pd.DataFrame(classification_report(y_test, y_pred_nb,output_dict = True))
result.loc[0]=[accuracy,scores.iloc[1,1],scores.iloc[2,1],scores.iloc[2,0],scores.iloc[0,1]]
```

```python
result
```

```python
mls = [SVM,lsvm,knn,lr,dt,rf,rfb,ab,nb]
result = pd.DataFrame(columns=['accuracy', 'precision', 'recall','specifity','f1'])
for i in range(len(mls)):
    mls[i].fit(X_ss_train, y_train)
    y_pred = mls[i].predict(X_ss_test)
    accuracy=accuracy_score(y_test, y_pred)
    print(mls[i])
    print(accuracy)
    print(confusion_matrix(y_test, y_pred))
    print('-----------------------------------------------------------')
    print(classification_report(y_test, y_pred))
    scores = pd.DataFrame(classification_report(y_test, y_pred,output_dict = True))
    result.loc[i]=[accuracy,scores.iloc[1,1],scores.iloc[2,1],scores.iloc[2,0],scores.iloc[0,1]]
```
